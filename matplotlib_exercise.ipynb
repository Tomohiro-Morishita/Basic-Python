```python
# 使うライブラリ
'''
python3.12以降を使用してる場合、japanize-matplotlibは使えません。
その場合、matplotlib-fontjaを使用してください。
'''


import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import matplotlib_fontja
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv("data/hist_data.csv")
targets = ["aデータ", "bデータ"]
fig, ax_1 = plt.subplots()
for target in targets:
    ax_1.hist(df_hist[target], bins=30, label=target, alpha=0.5)
ax_1.set_xlabel("温度")
ax_1.set_ylabel("度数")
ax_1.legend()
plt.title("aデータとbデータの温度に関する度数分布")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv("data/scatter_data.csv")
fig, ax_2 = plt.subplots()
color_map = {0:"r", 1:"b", 2:"g"}
for class_value, color in color_map.items():
    subset = df_scatter[df_scatter["class"]==class_value]
    ax_2.scatter(subset["length (cm)"], subset["width (cm)"], c=color, label=f"class{class_value}")
ax_2.legend()
ax_2.set_xlabel("length (cm)")
ax_2.set_ylabel("width (cm)")
plt.grid()
plt.title("クラス0～2のlengthとwidghに関する散布図")
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv("data/line_data.csv")
fig, ax_3 =plt.subplots()
ax_3.plot(df_line["期間"], df_line["件数"])
ax_3.set_xlabel("期間")
ax_3.set_ylabel("件数")

plt.xticks(rotation=45)
plt.grid()
plt.title("期間ごとの件数の推移")
```

```python
fig, ax_4 = plt.subplots()
ax_4.bar(df_line["期間"], df_line["費用"])
ax_3.set_xlabel("期間")
ax_3.set_ylabel("費用")

plt.xticks(rotation=45)
plt.grid()
plt.title("期間ごとの費用の推移")
```
